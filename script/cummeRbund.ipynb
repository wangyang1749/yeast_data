```r
cuff_data <- readCufflinks('diff_out',reBuild=T)
```

```r
# 不同转录本表达水平的密度分布
csDensity(genes(cuff_data))
```

```r
# 两种条件下转录本表达水平的情况
csScatter(genes(cuff_data),"EV","DNMT3B")
```

```r
## 火山图表示，不同条件下基因表达是否有显著性差异，
## 横坐标不同条件下基因表达水平的对数值
## 纵坐标为p检验中P值的负对数值
csVolcano(genes(cuff_data),"EV","DNMT3B", showSignificant=T,alpha=0.05)
```
